```python
def print_labels_counts_for_language(language: str):
    paths = get_paths(language=language)
    train_labels = (
        pd.read_csv(
            paths["train_labels"], sep="\t", encoding="utf-8", header=None
        )
            .rename(columns={0: "id", 1: "line", 2: "labels"})
            .set_index(["id", "line"])
    )

    train_labels['labels'] = train_labels['labels'].str.split(',')
    train_labels = train_labels.explode('labels')
    print(train_labels['labels'].value_counts())

languages_with_train = ["en", "fr", "ge", "it", "po", "ru"]
languages_without_train = ["es", "gr", "ka"]

# count the number of each label in dataset for all languages with train data
for language in languages_with_train:
    print("Language: ", language)
    print_labels_counts_for_language(language)
    print("-------------------------------------------------------------------------")
```

```python
def print_total_label_counts():
    "consider all the languages with train data, count the number of each label in dataset"
    total_labels = pd.DataFrame()
    for language in languages_with_train:
        paths = get_paths(language=language)
        train_labels = (
            pd.read_csv(
                paths["train_labels"], sep="\t", encoding="utf-8", header=None
            )
                .rename(columns={0: "id", 1: "line", 2: "labels"})
                .set_index(["id", "line"])
        )

        train_labels['labels'] = train_labels['labels'].str.split(',')
        train_labels = train_labels.explode('labels')
        total_labels = pd.concat([total_labels, train_labels])

    print(total_labels['labels'].value_counts())

    #print sum of all labels occurences
    print("Total labels: ", total_labels['labels'].value_counts().sum())
    
    #print number of rows in the dataset
    print("Total rows: ", total_labels.shape[0])

    #print number of rows without labels
    print("Rows without labels: ", total_labels[total_labels['labels'].isnull()].shape[0])

print_total_label_counts()
```
